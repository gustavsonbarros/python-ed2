```python
usuarios = {}

def menu():
    print("\n--- MENU ---")
    print("1 - Cadastrar novo usuario")
    print("2 - Verificar se o email foi cadastrado")
    print("3 - Remover Cadastro")
    print("4 - Listar todos os usuarios")
    print("5 - Sair")
    
while True:
    menu()
    opcao = input("Escolha a opcao: ")
    
    if opcao == "1":
        nome = input("Digite o nome do usuario: ")
        email = input("Digite o email do usuario: ")
        if email in usuarios:
            print("Este Email já está cadastrado.")
        else:
            usuarios[email] = nome
            print("Usuário cadastrado com sucesso.")
            
    elif opcao == "2":
        email = input("Digite o email para verificar: ")
        if email in usuarios:
            print(f"E-mail já cadastrado para o usuário {usuarios[email]}")
        else:
            print("E-mail ainda não cadastrado.")
            
    elif opcao == "3":
        email = input("Digite o e-mail para remover: ")
        if email in usuarios:
            del usuarios[email]
            print("Usuário removido com sucesso.")
        else:
            print("Este e-mail não está cadastrado.")
    
    elif opcao == "4":
        if usuarios:
            print("\nLista de usuários cadastrados:")
            for email, nome in usuarios.items():
                print(f"- {nome} ({email})")
        else:
            print("Nenhum usuário cadastrado ainda.")

    elif opcao == "5":
        print("Saindo do sistema...")
        break

    else:
        print("Opção inválida. Tente novamente.")
```
